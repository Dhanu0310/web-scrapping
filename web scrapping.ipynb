# importing libraries

```python

import requests
from bs4 import BeautifulSoup
```

```python
url="https://www.carandbike.com/research/top-10-best-cars-in-the-world"
```

# website response

```python

res=requests.get(url)
```

```python
res
```

# parsing

```python

soup=BeautifulSoup(res.text)
```

```python
soup
```

```python
cars=soup.select("tr+ tr td:nth-child(1) p")
```

```python
cars
```

```python
car_names=[]
```

```python

for item in cars:
    car_names.append(item.text)
    link="tr+ tr p"
```

```python
car_names
```

```python
speed=soup.select("tr+ tr td:nth-child(2) p")
```

```python
speed
```

```python
top_speed=[]
```

```python

for item in speed:
    top_speed.append(item.text)
```

```python
top_speed
```

```python

power=soup.select("tr+ tr td:nth-child(5) p")
```

```python
power
```

```python
high_power=[]
```

```python
for item in power:
    high_power.append(item.text)
```

```python
high_power
```

```python
quantity_torque=[]
```

```python

torque=soup.select("tr+ tr td:nth-child(4) p")
```

```python
torque
```

```python
for item in torque:
    quantity_torque.append(item.text)
```
